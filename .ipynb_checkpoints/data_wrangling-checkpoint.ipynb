### EDA

```python
df.Loan_Amount_Term.plot(kind="hist", colormap="RdGy")
```

```python
gender=df.groupby('Gender').count()['Loan_Status']
gender=pd.DataFrame(data=gender)
gender.reset_index(inplace=True)
gender.head()
```

```python
ax=gender.plot(kind='barh', colormap="RdGy")
ax.set_title('Loan Status Grouped by Gender')
ax.set_xlabel('Frequency')
ax.set_ylabel('Gender')
ax.set_yticklabels(['Women', 'Men'])
```

```python
gender.Loan_Status.dtype
```

### Observations
- the majority of the clientele is men
- we have unbalanced sample of approved versus rejected
- for women, there's almost a 50 50 split between accepted and rejected loans, whereas for men, there's more accepted than rejected loans
- the majority of accepted loans are for suburbans areas, and if you live in rural area, this will not likely be a determinant factor for wether you get approved or not verus if you live in suburban or urban areas
- the applicant and coapplicant income are negatively corr, and most of the obs are cluttered for low ranges of applicant + coapplicant incomes, the difference shows more at higher range, it would be probably worthwhile bucketing the observations
- the mean loan amount asked for is around 10-15k 
- the business is focused on >360 days loans
- the bank is currently targeting applicants who are at their early marriage lives, so more likely between 25-40 range
- although the graduates and non gradutes are not competing for the same loan amounts, they are competing for the same loan terms
- non self-employed people are asking for higher loan amounts compared to self-employed, so there might be a gap for the bank in terms of business loans?
- In the bad credit history category, you have more non self-employed compared to self-employed, maybe those people are trying get their finances in order
