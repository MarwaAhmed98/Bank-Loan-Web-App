```python
mdl=joblib.load('loan_model.pkl')
```

```python
df=pd.read_csv('production_data_loan.csv')
df=pd.get_dummies(df)
df.head()
```

```python
X=df.drop(['Loan_Status'], axis=1)
sc=MinMaxScaler()
X_test=sc.fit_transform(X)
X_test[1]
```

```python
y_pred=mdl.predict(X_test)
y_pred=(y_pred > 0.52)
data=pd.DataFrame(y_pred, columns=['Status'])
data=data.replace({True: 'Accepted', False: 'Rejected'})
data
```
