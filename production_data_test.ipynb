```python
import pickle
import numpy as np
from sklearn.externals import joblib
import pandas as pd
import tensorflow
from collections import Counter
```

```python
df=pd.read_csv('production_data_loan.csv')
df.drop('Loan_Status',axis=1, inplace=True)
```

```python
model=joblib.load('final_model.pkl')
sc=joblib.load('scaler_final.pkl')
```

```python
df=pd.get_dummies(df)
df
```

```python
X=sc.transform(df)
```

```python
y_pred=model.predict(X)
y_pred=(y_pred>0.50)
decision=pd.DataFrame(y_pred, columns=["Status"])
decision=decision.replace({True : "Accepted", False: "Rejected"})
list(decision.Status)
```
