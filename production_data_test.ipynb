```python
import pickle
import numpy as np
from sklearn.externals import joblib
import pandas as pd
import tensorflow
```

```python
df=pd.read_csv('production_data_loan.csv')
df.drop('Loan_Status',axis=1, inplace=True)
dum_X=pd.get_dummies(df)
dum_X.head()
```

```python
model=joblib.load('loan_model.pkl')
sc=joblib.load('scaler.pkl')
```

```python
X=sc.transform(dum_X)
```

```python
y_pred=model.predict(X)
y_pred=(y_pred>0.52)
decision=pd.DataFrame(y_pred, columns=["Status"])
decision=decision.replace({True : "Accepted", False: "Rejected"})
```
